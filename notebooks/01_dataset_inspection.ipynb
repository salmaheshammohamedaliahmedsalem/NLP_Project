# 01 Dataset Inspection

Inspect raw SemEval, RAID, and Arabic files before preprocessing. No model training happens here.

## Purpose

We inspect columns, sample rows, label distributions, missing values, duplicates, and possible leakage columns.

```python
from pathlib import Path
import pandas as pd

ROOT = Path("..").resolve()
RAW = ROOT / "data" / "raw"

def show_file(path):
    print("FILE", path)
    if path.suffix == ".csv":
        df = pd.read_csv(path)
    elif path.suffix in {".json", ".jsonl"}:
        df = pd.read_json(path, lines=path.suffix == ".jsonl")
    else:
        print("Unsupported preview", path)
        return
    display(df.head())
    print("shape", df.shape)
    print("columns", list(df.columns))
    print("missing")
    print(df.isna().sum().head(20))

for folder in ["semeval", "raid", "arabic"]:
    print("\n===", folder.upper(), "===")
    files = list((RAW / folder).glob("*"))
    print(files[:5])
    if files:
        show_file(files[0])
```

## Leakage Checklist

Do not use metadata as features: generator, model, source, attack, domain, prompt, dataset_name, ids, or label-derived columns.
